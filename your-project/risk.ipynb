```python
# Mise en place du jeu
def mise_en_place():
    
    # détermination du nombre de joueurs
    nb_joueurs_str = ""
    nb_joueurs_possibles = ["3", "4", "5", "6"]
    
    while nb_joueurs_str not in nb_joueurs_possibles:
        nb_joueurs_str = input (f"Nombre de joueurs (indiquez {min(nb_joueurs_possibles)} à {max(nb_joueurs_possibles)}) ? ")
    nb_joueurs = int(nb_joueurs_str)
    nb_armees = 50 - nb_joueurs * 5
    
    # constitution et mélange de la pioche
    pioche = list(DECK.keys())
    shuffle(pioche)

    # distribution des territoires aux joueurs
    territoires_joueurs = []
    armees_joueurs = []
    for j in range(nb_joueurs):
        territoires_joueurs.append({})

    # distribution des cartes alternativement à chaque joueur, tant qu'il en reste dans la pioche
    joueur = 0    
    while len(pioche) > 0:
        territoire = pioche.pop()
        # ajout du territoire au joueur s'il existe (il y a 2 joker dans la pioche, ne correspondant à aucun territoire)
        if territoire in LISTE_PAYS:
            territoires_joueurs[joueur][territoire] = 1
        # joueur suivant
        joueur = (joueur + 1) % nb_joueurs
        
    # calcul du nombre d'armées restantes par joueur    
    for j in range(nb_joueurs):
        armees_joueurs.append(nb_armees - sum(territoires_joueurs[j].values()))   
        
    return nb_joueurs, armees_joueurs, territoires_joueurs


```

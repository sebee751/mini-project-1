```python
from random import shuffle, randint

# symboles associés aux joueurs
LISTE_ICONES_JOUEURS = ["*", "$", "%", "o", "x", "^"]

# Les cartes du jeu
DECK = {"AFG":["Afghanistan", "Asie", "Fantassin"], 
        "AFN":["Afrique du Nord", "Afrique", "Fantassin"], 
        "AFS":["Afrique du Sud", "Afrique", "Canon"], 
        "AFO":["Afrique Orientale", "Afrique", "Canon"], 
        "ASK":["Alaska", "Am. du N.", "Fantassin"], 
        "ALB":["Alberta", "Am. du N.", "Fantassin"], 
        "AMC":["Amérique Centrale", "Am. du N.", "Cavalier"], 
        "ARG":["Argentine", "Am. du S.", "Fantassin"], 
        "AOC":["Australie Occidentale", "Océanie", "Canon"], 
        "AOR":["Australie Orientale", "Océanie", "Fantassin"], 
        "BRE":["Brésil", "Am. du S.", "Canon"], 
        "CHN":["Chine", "Asie", "Cavalier"], 
        "CGO":["Congo", "Afrique", "Cavalier"], 
        "EGY":["Égypte", "Afrique", "Fantassin"], 
        "EDE":["État de l'Est", "Am. du N.", "Canon"], 
        "EDO":["État de l'Ouest", "Am. du N.", "Fantassin"], 
        "EUN":["Europe du Nord", "Europe", "Cavalier"], 
        "EUS":["Europe du Sud", "Europe", "Cavalier"], 
        "EUO":["Europe Occidentale", "Europe", "Fantassin"], 
        "GBR":["Grande-Bretagne", "Europe", "Cavalier"], 
        "GRO":["Groënland", "Am. du N.", "Cavalier"], 
        "IND":["Inde", "Asie", "Fantassin"], 
        "IDN":["Indonésie", "Asie", "Cavalier"], 
        "ISL":["Islande", "Europe", "Fantassin"], 
        "JPN":["Japon", "Asie", "Fantassin"], 
        "KAM":["Kamchatka", "Asie", "Cavalier"], 
        "MDG":["Madagascar", "Afrique", "Fantassin"], 
        "MNG":["Mongolie", "Asie", "Canon"], 
        "MYO":["Moyen-Orient", "Asie", "Canon"], 
        "NVG":["Nouvelle Guninée", "Océanie", "Cavalier"], 
        "ONT":["Ontario", "Am. du N.", "Cavalier"], 
        "OUR":["Oural", "Asie", "Cavalier"], 
        "PER":["Pérou", "Am. du S.", "Cavalier"], 
        "QBC":["Québec", "Am. du N.", "Canon"], 
        "SCD":["Scandinavie", "Europe", "Canon"], 
        "SIA":["Siam", "Asie", "Canon"], 
        "SIB":["Sibérie", "Asie", "Canon"], 
        "TCH":["Tchita", "Asie", "Fantassin"], 
        "TNO":["Territoire du Nord-Ouest", "Am. du N.", "Canon"], 
        "UKR":["Ukraine", "Europe", "Canon"], 
        "VNZ":["Vénézuela", "Am. du S.", "Canon"], 
        "YAK":["Yakoutie", "Asie", "Cavalier"], 
        "JK1":["", "", "Joker"], 
        "JK2":["", "", "Joker"]
       }

# la carte du monde
CARTE = ["ASK────────────────────────────────┐     ",
         "│└─┐                        YAK───┐│     ",
         "│  │                        / \   ││     ",
         "│ TNO-GRO────────┐        SIB-TCH-KAM    ",
         "│ / \ / \        │        /│\ /   ││     ",
         "ALB-ONT-QBC     ISL     OUR│MNG───┘│     ",
         "  \ / \ /       / \     ││\│/ \┌───┘     ",
         "  EDO-EDE     GBR-SCD   ││CHN JPN        ",
         "    \ /       / \ / \┌──┘│/│\            ",
         "    AMC     EUO-EUN-UKR-AFG│SIA-IDN      ",
         "     |       ││  |   │   │\│/   / \      ",
         "    VNZ      │└─EUS──┤   │IND AOC-NVG    ",
         "    / \      └─┐/ \  └──┐│/     \ /      ",
         "  PER-BRE─────AFN-EGY───MYO     AOR      ",
         "    \ /       / \ /                      ",
         "    ARG     CGO-AFO                      ",
         "              \ / \                      ",
         "              AFS-MDG                    "
        ]

# les frontières entre tous les pays
FRONTIERES = [("ASK", "ALB"), ("ASK", "KAM"), ("ASK", "TNO"), ("YAK", "SIB"), ("YAK", "TCH"), ("YAK", "KAM"), ("TNO", "GRO"), ("TNO", "ALB"), 
              ("TNO", "ONT"), ("GRO", "ISL"), ("GRO", "ONT"), ("GRO", "QBC"), ("SIB", "OUR"), ("SIB", "CHN"), ("SIB", "MNG"), ("SIB", "TCH"), 
              ("TCH", "MNG"), ("TCH", "KAM"), ("KAM", "MNG"), ("KAM", "JPN"), ("ALB", "EDO"), ("ALB", "ONT"), ("ONT", "EDO"), ("ONT", "EDE"), 
              ("ONT", "QBC"), ("QBC", "EDE"), ("ISL", "GBR"), ("ISL", "SCD"), ("OUR", "UKR"), ("OUR", "AFG"), ("OUR", "CHN"), ("MNG", "CHN"), 
              ("MNG", "JPN"), ("EDO", "AMC"), ("EDO", "EDE"), ("EDE", "AMC"), ("GBR", "EUO"), ("GBR", "EUN"), ("GBR", "SCD"), ("SCD", "EUN"), 
              ("SCD", "UKR"), ("CHN", "AFG"), ("CHN", "IND"), ("CHN", "SIA"), ("AMC", "VNZ"), ("EUO", "AFN"), ("EUO", "EUS"), ("EUO", "EUN"), 
              ("EUN", "EUS"), ("EUN", "UKR"), ("UKR", "EUS"), ("UKR", "MYO"), ("UKR", "AFG"), ("AFG", "MYO"), ("AFG", "IND"), ("SIA", "IND"), 
              ("IDN", "AOC"), ("IDN", "NVG"), ("VNZ", "PER"), ("VNZ", "BRE"), ("EUS", "AFN"), ("EUS", "EGY"), ("IND", "MYO"), ("AOC", "AOR"), 
              ("AOC", "NVG"), ("NVG", "AOR"), ("PER", "ARG"), ("PER", "BRE"), ("BRE", "ARG"), ("BRE", "AFN"), ("AFN", "CGO"), ("AFN", "AFO"), 
              ("AFN", "EGY"), ("EGY", "AFO"), ("EGY", "MYO"), ("CGO", "AFS"), ("CGO", "AFO"), ("AFO", "AFS"), ("AFO", "MDG"), ("AFS", "MDG")]

# restitution de l'ensemble des frontières de chaque pays sous forme de dictionnaire
LISTE_PAYS = set([pays1 for pays1, pays2 in FRONTIERES]+[pays2 for pays1, pays2 in FRONTIERES])

PAYS_FRONTIERES = {pays: [] for pays in LISTE_PAYS}
for pays1, pays2 in FRONTIERES:
    PAYS_FRONTIERES[pays1].append(pays2)
    PAYS_FRONTIERES[pays2].append(pays1)

    
    
# Mise en place du jeu
def mise_en_place():
    
    # détermination du nombre de joueurs
    nb_joueurs_str = ""
    nb_joueurs_possibles = ["3", "4", "5", "6"]
    
    while nb_joueurs_str not in nb_joueurs_possibles:
        nb_joueurs_str = input (f"Nombre de joueurs (indiquez {min(nb_joueurs_possibles)} à {max(nb_joueurs_possibles)}) ? ")
    nb_joueurs = int(nb_joueurs_str)
    
    nb_armees = 50 - nb_joueurs * 5
    
    # constitution et mélange de la pioche
    pioche = list(DECK.keys())
    shuffle(pioche)

    # distribution des territoires aux joueurs
    territoires_joueurs = []
    armees_joueurs = []
    for j in range(nb_joueurs):
        territoires_joueurs.append({})

    # distribution des cartes alternativement à chaque joueur, tant qu'il en reste dans la pioche
    joueur = 0    
    while len(pioche) > 0:
        territoire = pioche.pop()
        # ajout du territoire au joueur s'il existe (il y a 2 joker dans la pioche, ne correspondant à aucun territoire)
        if territoire in LISTE_PAYS:
            territoires_joueurs[joueur][territoire] = 1
        # joueur suivant
        joueur = (joueur + 1) % nb_joueurs
        
    # calcul du nombre d'armées restantes par joueur    
    for j in range(nb_joueurs):
        armees_joueurs.append(nb_armees - sum(territoires_joueurs[j].values()))   
        
    return nb_joueurs, armees_joueurs, territoires_joueurs


'''
continents = set([DECK[cle][1] for cle in DECK if DECK[cle][1] != ''])
liste_pays_continents = [f"- {continent} : "+", ".join([pays for pays in DECK if DECK[pays][1]==continent]) for continent in continents]

legende = ["LÉGENDE :", "---------", "", "Continents : "] + sorted(liste_pays_continents)
legende
'''



def mise_en_forme_ligne(ligne, remplacements):
    for caractere in remplacements.keys():
        ligne = ligne.replace(caractere, remplacements[caractere])
    return ligne


def afficher_carte(territoires_joueurs):
    occupation_territoires = {}      
    
    for j in range(len(territoires_joueurs)):
        
        for cle in territoires_joueurs[j]:
            nb_unites = territoires_joueurs[j][cle]
            
            if nb_unites > 99:
                str_nb_unites = "++"
            elif nb_unites <= 9:
                str_nb_unites = f"0{str(nb_unites)}"
            else: 
                str_nb_unites = str(nb_unites)
                
            occupation_territoires[cle] = f"{LISTE_ICONES_JOUEURS[j]}{str_nb_unites}"

        remplacements = {"-":" ", "─":" ", "└":" ", "┘":" ", "┐":"│", "┌":"│", "┤":"│", "├":"│"}
        
    for ligne in CARTE:
        
        ligne_occupation = mise_en_forme_ligne(ligne, occupation_territoires)
        print(ligne)
        if ligne != ligne_occupation:
            print(mise_en_forme_ligne(ligne_occupation, remplacements))

            
def placer_armees(armees_joueurs, territoires_joueurs):
    joueur = 0
    nb_joueurs = len(armees_joueurs)
    
    while sum(armees_joueurs) > 0:
        joueur_suivant = (joueur + 1) % nb_joueurs
        nb_armees = armees_joueurs[joueur]
        if nb_armees > 0: 
            entree = input(f"Joueur '{LISTE_ICONES_JOUEURS[joueur]}', il vous reste {nb_armees} armées. \nIndiquez le pays et le nombre d'unités à placer : ").upper()
            if entree.upper() == 'QUIT':
                return False
            if len(entree.split(' ')) == 2:
                pays, nombre = entree.split()
                try: 
                    nombre = int(nombre)
                except:
                    nombre = 0
                    print(f"Nombre d'armées invalide.")
            else:
                pays = ''
                print(f"Pays sur 3 caractères, espace, nombre d'unités (ex. : {list(territoires_joueurs[joueur].keys())[0]} 1)")
            if pays in territoires_joueurs[joueur].keys() and nombre > 0:
                nb_deplacements = min(nb_armees, nombre)
                territoires_joueurs[joueur][pays] += nb_deplacements
                armees_joueurs[joueur] = nb_armees - nb_deplacements
                joueur = joueur_suivant
                afficher_carte(territoires_joueurs)
            else :
                if pays != '':
                    print(f"Le territoire '{pays}' n'est pas occupé par vos armées.")
        else:
            joueur = joueur_suivant
            
    return territoires_joueurs
    
# def deplacer_armee():

def combattre(joueur, territoires_joueurs):
    attaque = ""
    while attaque.upper() != 'STOP':
        attaque = input(f"Joueur '{LISTE_ICONES_JOUEURS[joueur]}', indiquez le territoire de depuis lequel vous lancez l'attaque, le territoire attaqué et le nombre d'armées. \nTapez 'stop' si vous ne souhaitez plus lancer d'attaques : ")
        if attaque.upper() == 'STOP':
            return territoires_joueurs
        else:
            if len(attaque.split()) == 3:
                territoire_origine, territoire_destination, nb_armees = attaque.split()
                territoire_origine = territoire_origine.upper()
                territoire_destination = territoire_destination.upper()
                try:
                    nb_armees_attaque = int(nb_armees)
                except:
                    nb_armees_attaque = 0
                    print("Nombre d'armées invalide.")
            else:
                territoire_origine = ""
                territoire_destination = ""
                print("Nombre de paramètres incorrect.")
                
        if territoire_origine in territoires_joueurs[joueur].keys() and territoire_destination in LISTE_PAYS:
            total_armees = territoires_joueurs[joueur][territoire_origine]
            [joueur_defense] = [joueur for joueur, liste_territoires in enumerate(territoires_joueurs) if territoire_destination in liste_territoires]
            
            if nb_armees_attaque <= 0:
                print("Nombre d'armées invalide.")
            elif nb_armees_attaque > min(3, total_armees - 1):
                print("Vous ne pouvez pas attaquer avec plus de 3 armees. Au moins une armée du territoire d'origine de l'attaque ne doit pas être impliquée dans le combat")
            elif territoire_destination not in PAYS_FRONTIERES[territoire_origine]:
                print("Vous ne pouvez pas attaquer ce territoire depuis le territoire d'origine de votre attaque.")
            elif joueur == joueur_defense:
                print("Vous ne pouvez pas attaquer un territoire que vous occupez déjà.")
            else:
                if territoires_joueurs[joueur_defense][territoire_destination] > 1:
                    nb_armees_defense = ""
                    while nb_armees_defense not in [1, 2]:
                        try:
                            nb_armees_defense = int(input(f"Joueur '{LISTE_ICONES_JOUEURS[joueur_defense]}', souhaitez vous défendre avec 1 ou 2 armées ? : "))
                        except:
                            pass
                else:
                    nb_armees_defense = 1

                des_attaque = sorted([randint(1,6) for i in range(nb_armees_attaque)])
                des_defense = sorted([randint(1,6) for i in range(nb_armees_defense)])

                resultat_des = f"Tirs de dés : \n- Joueur '{LISTE_ICONES_JOUEURS[joueur]}' : {des_attaque}, \n- Joueur '{LISTE_ICONES_JOUEURS[joueur_defense]}' : {des_defense}."

                pertes_attaque = 0
                pertes_defense = 0
                for de in range(min(len(des_attaque), len(des_defense))):
                    if des_attaque.pop() > des_defense.pop():
                        pertes_defense += 1
                    else:
                        pertes_attaque += 1

                resultat_pertes = f"Pertes : \n- Joueur '{LISTE_ICONES_JOUEURS[joueur]}' : {pertes_attaque}, \n- Joueur '{LISTE_ICONES_JOUEURS[joueur_defense]}' : {pertes_defense} "

                # résolution du combat
                
                nb_survivants_attaque = nb_armees_attaque - pertes_attaque
                
                territoires_joueurs[joueur][territoire_origine] -= pertes_attaque
                territoires_joueurs[joueur_defense][territoire_destination] -= pertes_defense
                
                afficher_carte(territoires_joueurs)
                print(resultat_des)
                print(resultat_pertes)
                
                if territoires_joueurs[joueur_defense][territoire_destination] == 0:
                    del territoires_joueurs[joueur_defense][territoire_destination]
                    territoires_joueurs[joueur][territoire_origine] -= nb_survivants_attaque
                    territoires_joueurs[joueur][territoire_destination] = nb_survivants_attaque
                    
                    afficher_carte(territoires_joueurs)
                    print(resultat_des)
                    print(resultat_pertes)
                    
                    if territoires_joueurs[joueur][territoire_origine] > 1:
                        armees_deplacees = ""
                        nb_min = 0
                        nb_max = territoires_joueurs[joueur][territoire_origine]
                        while armees_deplacees not in range(nb_min, nb_max):
                            try:
                                armees_deplacees = int(input(f"Combien d'autres armées souhaitez-vous déplacer ({nb_min} à {nb_max - 1})? "))
                            except:
                                pass
 
                        territoires_joueurs[joueur][territoire_origine] -= armees_deplacees
                        territoires_joueurs[joueur][territoire_destination] += armees_deplacees
        
                        afficher_carte(territoires_joueurs)
                
            
        else:
            if territoire_origine != "":
                print("Vous n'occupez pas le territoire d'origine de l'attaque, ou le territoire que vous attaquez n'existe pas.")


def tour_de_jeu(joueur, territoires_joueurs):
        
    # renforts : nb territoires//3, 3 mini. à développer : échange de cartes, nb / continent (Oc:2 AS:2 Af:3 Eu:5 AN:5 As:7) /
    armees_joueurs = [0, 0, 0, 0, 0, 0]
    nb_territoires = len(territoires_joueurs[joueur])
    armees_joueurs[joueur] = max(3, nb_territoires // 3)
    
    territoires_joueurs = placer_armees(armees_joueurs, territoires_joueurs)
    if not territoires_joueurs:
        return False
    territoires_joueurs = combattre(joueur, territoires_joueurs)
    
    return territoires_joueurs
```
